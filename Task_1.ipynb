```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: ages of a population
ages = np.random.randint(0, 100, size=1000)  # Generate 1000 random ages between 0 and 100

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution in a Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.xticks(np.arange(0, 101, 10))  # Set x-ticks from 0 to 100
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data: gender distribution
genders = ['Male', 'Female', 'Non-binary']
counts = [500, 400, 100]  # Example counts for each gender

# Create a bar chart
plt.figure(figsize=(8, 5))
plt.bar(genders, counts, color=['blue', 'pink', 'purple'])
plt.title('Gender Distribution in a Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
